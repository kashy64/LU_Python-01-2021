## 1) Write a Python function to find the Max of three numbers.

```python
## First solution
```

```python
def max_of_two( x, y ):
    if x > y:
        return x
    return y
def max_of_three( x, y, z ):
    return max_of_two( x, max_of_two( y, z ) )
print(max_of_three(155,50,25))
```

```python
## Second Solution
```

```python
def max_three(x,y,z):
    if x > y:
        return x
    if y > z:
        return y
    if z > x:
        return z
max_three(155,50,25)
```

## 2) Write a Python function that checks whether a passed string is palindrome or not

```python
def Palindrome(s):
    if s==s[::-1]:
        print("It is a Palindrome")
    else:
        print("It isn't a Palindrome")
Palindrome("malayalam")
```

## 3) Write a Python function that accepts a string and calculate the number of uppercase letters and lowercase letters

```python
def up_low(s):      
    u = sum(1 for i in s if i.isupper())
    l = sum(1 for i in s if i.islower())
    print( "No. of Upper case characters : %s" %u)
    print( "No. of Lower case characters : %s" %l)

up_low("Hello Mr. Rogers, how are you this fine Tuesday?")
```

## 4) Write a Python function to sum all the numbers in a list

```python
def sum_all(list):
    return sum(list)
sum_all([100,200,300])
```

## 5) Write a Python function to multiply all the numbers in a list

```python
def mul_all(l):
    a=1
    for i in l:
        a*=i
    return a
mul_all([1,2,3])
```

## 6) Write a Python function that takes a list and returns a new list with unique elements of the first list

```python
def unique(l):  
    l_s = set(l) 
    u = (list(l_s)) 
    for x in u: 
        print(x)
unique([10,20,25,20,25,10,50])
```
